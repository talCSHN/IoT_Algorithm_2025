## 2번

````python
# 전역변수 선언
SIZE = 5
queue = [None for _ in range(SIZE)]
front = -1
rear = -1

# 큐가 가득찼는지 판별
def isQueueFull():
    # 전역변수 사용
    global SIZE, queue, front, rear

    # 맨 뒤 데이터가 큐의 마지막에 위치할 때
    if rear == SIZE - 1:
        return True
    else:
        return False

# 큐가 비었는지 판별
def isQueueEmpty():
    # 전역변수 사용
    global SIZE, queue, front, rear

    # 맨 앞 데이터가 맨 뒤 일 때
    if front == rear :
        return True
    else:
        return False

# 큐에 데이터 삽입
def enQueue(data):
    # 전역변수 사용
    global SIZE, queue, front, rear
    # 큐가 가득 찼는지 판별
    if isQueueFull():
        print('큐가 가득 찼음')
        return  # 조건문 종료
    rear += 1
    queue[rear] = data  # 큐의 끝에 data 삽입

# 큐에서 데이터 빼기
def deQueue():
    # 전역변수 사용
    global SIZE, queue, front, rear
    # 큐가 비었는지 판별
    if isQueueEmpty():
        print('큐가 비었음')
        return  # 조건문 종료
    
    front += 1
    data = queue[front] # 큐 맨앞 데이터를 data에 할당
    queue[front] = None # 큐 맨앞에 None 할당

    # 한칸씩 앞으로 당기기
    for i in range(front+1, rear+1):
        queue[i-1] = queue[i]
        queue[i] = None
    # front는 -1로 초기화해주고 rear는 -1해줌
    front = -1
    rear -= 1

    return data

# 큐 맨 앞의 값 찾기
def peek():
    # 전역변수 사용
    global SIZE, queue, front, rear

    # 큐가 비었는지 판별
    if isQueueEmpty():
        print('큐가 비었음')
        return None # None 반환

    # 안비었으면 queue 맨앞의 값 반환
    print(f'맨 앞 대기 손님 : {queue[front+1]}')
    # return queue[front+1]
    
def run():

    # 전역변수 사용
    global SIZE, queue, front, rear

    while True:
        inputNum = int(input('번호 입력(1:대기손님 등록, 2:맨 앞 대기손님 입장, 3:맨 앞 대기손님, 4:종료)'))
        # 1번 입력시 대기손님 등록
        if inputNum == 1:
            waitingName = input('대기 손님 이름 입력')
            enQueue(waitingName)
            print(f'대기 줄 상태 : {queue}')
        # 2번 입력시 대기손님 입장 -> 큐에서 제거
        elif inputNum == 2:
            print(f'{queue[front+1]}님 식당에 들어감')
            deQueue()
            print(f'대기 줄 상태 : {queue}')
        # 맨 앞 대기손님 확인
        elif inputNum == 3:
            peek()
        else:
            return
# 실행
if __name__ == '__main__':
    run()
    print('식당 영업 종료')
````

## 3번

````python
import random

# TreeNode 클래스 선언
class TreeNode():
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None

memory = []
root = None
dataArr = ['콘말차', '삿포로 맥주', '아카페라 벤티 헤이즐넛', 
           '레어치즈푸딩', '척오리지널 메가 사워', '요아정 요거트컵', '페퍼로니피자주먹밥', 
           '널담 슈톨렌', '딸기마시멜로케이크', '버니공쥬 딸기뚱카롱', '고추잡채호빵', '체다 슈크림붕어스낵']
# dataArr 내에서 물품을 랜덤하게 20개 꺼냄(중복 가능)
sellArr = [random.choice(dataArr) for _ in range(20)]
print(f'오늘 판매된 물건 : {sellArr}')

def preorder(node):
    if node == None:
        return
    print(node.data, end='\n')
    preorder(node.left)
    preorder(node.right)

print('오늘 판매된 종류(중복 제거됨) ->  ', end=' ')
# preorder(root)

if __name__ == '__main__':

    # node객체 생성
    node = TreeNode()
    # node의 데이터를 sellArr 첫번째 값으로 초기화
    node.data = sellArr[0]
    # root에 node 할당
    root = node
    # memory 리스트에 node 삽입
    memory.append(node)

    # 반복문으로 데이터 할당(sellArr의 두번째 인덱스부터 마지막까지)

    for name in sellArr[1:]:
        node = TreeNode()
        node.data = name
        
        # 현재 데이터에 root 할당
        current = root
        while True:
            if name == current.data: break
            # current의 데이터보다 작으면
            if name < current.data:
                if current.left ==None:
                    current.left = node # current의 왼쪽 링크에 node 할당
                    memory.append(node)
                    break
                current = current.left
            else:
                if current.right == None:
                    current.right = node
                    memory.append(node)
                    break
                current = current.right

    print('#####')
    preorder(root)

    findName = input('찾을 이름 입력 > ')

    current = root  # 루트노드부터 검색 시작
    count = 0
    while True:
        if findName == current.data:
            print(f'{findName} 찾음')
            count += 1
            break
        elif findName < current.data:
            if current.left == None:
                print(f'{findName} 트리에 없음')
                break
            else:
                current = current.left
                count += 1
        else:
            if current.right == None:
                print(f'{findName} 트리에 없음')
                break
            else:
                current = current.right
                count += 1

    print(f'검색 종료. 총 검색 횟수:{count}')

    # 트리노드 삭제(리버풀, 바르샤 삭제)
    delName = input('삭제할 노드 입력 > ')

    current = root
    parent = None
    while True:
        if delName == current.data: # 삭제할 때

            if current.left == None and current.right == None:  # 리프노드
                if parent.left == current:
                    parent.left = None
                else:
                    parent.right = None
                del(current)
            elif current.left != None and current.right == None: # 왼쪽 노드에 자식이 있고, 오른쪽에는 없을때
                if parent.left == current:
                    parent.left = current.left
                else:
                    parent.right = current.left
                del(current)
            elif current.left == None and current.right != None: # 오른쪽 노드에 자식이 있고, 왼쪽에는 없을때
                if parent.left == current:
                    parent.left = current.right
                else:
                    parent.right = current.right
                del(current)
            print(f'{delName} 삭제함')
            break
        elif delName < current.data:    # 삭제할 데이터가 트리 왼쪽에 있음
            if current.left == None:
                print(f'{delName} 트리에 없음')
                break
            else:
                parent = current
                current = current.left
        else:   # 삭제할 데이터가 트리 오른쪽에 있으면
            if current.right == None:
                print(f'{delName} 트리에 없음')
                break
            else:
                parent = current
                current = current.right

````

````python
import random

## 이진검색 함수 정의
def binSearch(arr, data):
    pos = -1
    start = 0
    end = len(arr) - 1  # 리스트의 마지막 인덱스

    global count
    while start <= end: # 시작이 끝보다 커지면 검색 실패
        mid = (start + end) // 2
        if data == arr[mid]:
            count += 1
            return mid
        elif data > arr[mid]:
            start = mid + 1 # 뒤쪽에 데이터가 있다
            count += 1
        else:
            end = mid - 1   # 앞쪽에 데이터가 있다
            count += 1
    return pos

# 변수
dataArr = [random.randint(0, 100000) for _ in range(100000)]
dataArr.sort()
foundData = random.choice(dataArr)
count = 0
# dataArr = [1,2,3,4,5]
print(f'배열 -> {dataArr}')

# foundData = int(input('찾을 값 입력(숫자) '))

pos = binSearch(dataArr, foundData)
if pos == -1:
    print(f'{foundData}(이)가 없음')
else:
    print(f'{foundData}는 {pos}에 존재')

print(f'위치값 : {pos}')
print(f'검색 종료. 총 검색 횟수:{count}')
````
